## Live Sweep Viewer
Re-run the cell below to see the latest sweep from `mim_moku.py`.

```python
import json
import numpy as np
import matplotlib.pyplot as plt

with open('latest_sweep.json') as f:
    d = json.load(f)

freq = np.array(d['frequency'])
gain = np.array(d['gain_db'])

fig, ax = plt.subplots(figsize=(12, 5))
ax.semilogx(freq, gain, linewidth=0.8)
ax.axvline(d['resonant_hz'], color='r', linestyle='--', linewidth=1.2,
           label=f"peak = {d['resonant_hz']:.0f} Hz")
ax.set_xlabel('Frequency (Hz)', fontsize=12)
ax.set_ylabel('Gain (dB)', fontsize=12)
ax.set_title(f"Run {d['run']}  —  T = {d['temperature_C']:.1f} °C  —  C = {d['capacitance_pF']:.2f} pF",
             fontsize=13)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
